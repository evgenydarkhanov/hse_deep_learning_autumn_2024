```python
class MyClass:
    def __init__(self, name):
        self._name = name

    def my_method(self, arg):
        print(f"hello {arg}!")

    @property
    def name(self):
        """ docstring """
        return self._name


my_class = MyClass("planet")
my_class.my_method("world")
print(my_class.name)
```
